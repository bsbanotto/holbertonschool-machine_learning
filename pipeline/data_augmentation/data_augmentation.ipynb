```python
# Imports
import numpy as np
import tensorflow as tf

import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
```

```python
# Task 0. Flip
def flip_image(image):
    """
    Function to flip an image horizontally

    Args:
        image: 3D tf.Tensor containing the image to filp

    Returns:
        the filpped image
    """
    return tf.image.flip_left_right(image)
```

```python
# Before Flipping

tf.compat.v1.enable_eager_execution()
tf.compat.v1.set_random_seed(0)

doggies = tfds.load('stanford_dogs', split='train', as_supervised=True)
for image, _ in doggies.shuffle(10).take(1):
    plt.imshow(image)
    plt.show()
```

```python
# 0-main
tf.compat.v1.enable_eager_execution()
tf.compat.v1.set_random_seed(0)

doggies = tfds.load('stanford_dogs', split='train', as_supervised=True)
for image, _ in doggies.shuffle(10).take(1):
    plt.imshow(flip_image(image))
    plt.show()
```

```python
# Task 1. Crop
def crop_image(image, size):
    """
    Performs a random crop of an image

    Args:
        image: 3D tf.Tensor containing the image to crop
        size: tuple containing the size of the crop

    Returns:
        the cropped image
    """
    return tf.image.random_crop(image, size)
```

```python
# Before cropping
tf.compat.v1.enable_eager_execution()
tf.compat.v1.set_random_seed(1)

doggies = tfds.load('stanford_dogs', split='train', as_supervised=True)
for image, _ in doggies.shuffle(10).take(1):
    plt.imshow(image)
    plt.show()
```

```python
# 1-main
tf.compat.v1.enable_eager_execution()
tf.compat.v1.set_random_seed(1)

doggies = tfds.load('stanford_dogs', split='train', as_supervised=True)
for image, _ in doggies.shuffle(10).take(1):
    plt.imshow(crop_image(image, (200, 200, 3)))
    plt.show()
```

```python
# Task 2. Rotate
def rotate_image(image):
    """
    Rotates an image by 90 degrees, counter-clockwise

    Args:
        image: 3D tf.Tensor containing the image to rotate

    Returns:
        the rotated image
    """
    return tf.image.rot90(image)
```

```python
# Before rotating
tf.compat.v1.enable_eager_execution()
tf.compat.v1.set_random_seed(2)

doggies = tfds.load('stanford_dogs', split='train', as_supervised=True)
for image, _ in doggies.shuffle(10).take(1):
    plt.imshow(image)
    plt.show()
```

```python
# 2-main
tf.compat.v1.enable_eager_execution()
tf.compat.v1.set_random_seed(2)

doggies = tfds.load('stanford_dogs', split='train', as_supervised=True)
for image, _ in doggies.shuffle(10).take(1):
    plt.imshow(rotate_image(image))
    plt.show()
```

```python
# Task 3. Shear
def shear_image(image, intensity):
    """
    Function that randomly shears an image

    Args:
        image: 3D tf.Tensor containing the image to shear
        intensity: the intensity with which the image should be sheared

    Returns:
        the sheared image
    """
    return tf.keras.preprocessing.image.random_shear(image,
                                                     intensity,
                                                     row_axis=1,
                                                     col_axis=0,
                                                     channel_axis=2,
                                                     fill_mode='nearest',
                                                     cval=0.0,
                                                     interpolation_order=1
                                                     )
```

```python
# Before Shear
tf.compat.v1.enable_eager_execution()
tf.compat.v1.set_random_seed(3)

doggies = tfds.load('stanford_dogs', split='train', as_supervised=True)
for image, _ in doggies.shuffle(10).take(1):
    plt.imshow(image)
    plt.show()
```

```python
# 3-main
from tensorflow.python.ops.numpy_ops import np_config
np_config.enable_numpy_behavior()

tf.compat.v1.enable_eager_execution()
tf.compat.v1.set_random_seed(3)

doggies = tfds.load('stanford_dogs', split='train', as_supervised=True)
for image, _ in doggies.shuffle(10).take(1):
    plt.imshow(shear_image(image=image, intensity=50))
    plt.show()
```

```python
# Task 4. Brightness
def change_brightness(image, max_delta):
    """
    Function that randomly changes the brightness of an image

    Args:
        image: 3D tf.Tensor containing the image to change
        max_delta: the maximum amount the image should be brightened

    Returns:
        Brightened image
    """
    return tf.image.random_brightness(image, max_delta)
```

```python
# Before Brightness Change
tf.compat.v1.enable_eager_execution()
tf.compat.v1.set_random_seed(4)

doggies = tfds.load('stanford_dogs', split='train', as_supervised=True)
for image, _ in doggies.shuffle(10).take(1):
    plt.imshow(image)
    plt.show()
```

```python
# 4-main
tf.compat.v1.enable_eager_execution()
tf.compat.v1.set_random_seed(4)

doggies = tfds.load('stanford_dogs', split='train', as_supervised=True)
for image, _ in doggies.shuffle(10).take(1):
    plt.imshow(change_brightness(image, 0.3))
    plt.show()
```

```python
# Task 5. Hue
def change_hue(image, delta):
    """
    Function that changes the hue of an image

    Args:
        image: 3D tf.Tensor containing the image to be changed
        delta: amount the hue should change

    Returns:
        Changed image
    """
    return tf.image.adjust_hue(image, delta)
```

```python
# Before hue change
tf.compat.v1.enable_eager_execution()
tf.compat.v1.set_random_seed(5)

doggies = tfds.load('stanford_dogs', split='train', as_supervised=True)
for image, _ in doggies.shuffle(10).take(1):
    plt.imshow(image)
    plt.show()
```

```python
# 5-main
tf.compat.v1.enable_eager_execution()
tf.compat.v1.set_random_seed(5)

doggies = tfds.load('stanford_dogs', split='train', as_supervised=True)
for image, _ in doggies.shuffle(10).take(1):
    plt.imshow(change_hue(image, -0.5))
    plt.show()
```
